# Mergesort

The mergesort algorithm belongs to a class of algorithms that operate by breaking the problem into smaller pieces that are easier to manage. This makes *Mergesort* a good candidate for a recursive solution:

1. Is `len(lst) == 1`?
    * Yes: return lst
2. `mid = len(lst)/2`
3. `left = mergesort(lst[:mid])`
4. `right = mergesort(lst[mid:])`
5. `return merge(left, right)`

where `merge` goes through both lists and combines them together. The `left` and `right` input to `merge` is guaranteed to be in order, so `merge` only has to alternate between the two inputs to create a combined version:

1. Is `left[i] > right[j]`?
    * Yes: `result[k++] = left[i++]`
    * No: `result[k++] = right[j++]`
2. `return result`


The code for `mergesort` follows the sketch we made above:

```python
def mergesort(lst):
    if len(lst) == 1: return lst
    
    mid = len(lst)/2
    left = mergesort(lst[:mid])
    right = mergesort(lst[mid:])
    return merge(left, right)
```

The code for `merge` also follows the sketch made above, though we need to include more checks for boundary conditions to deal with the cases of one list becoming empty before the other:

```python
def merge(left, right):
    i = j = 0
    
    result = []
    while i < len(left) or j < len(right):
        if i >= len(left):
            result.append(right[j])
            j += 1            
        elif j >= len(right):
            result.append(left[i])
            i += 1            
        elif left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            
    return result
```
